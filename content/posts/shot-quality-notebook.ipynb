# A short notebook on shot-quality distribution

Two teams can finish a match with identical xG totals and have produced radically different evenings. This notebook demonstrates why the *shape* of a team's shot-quality distribution matters at least as much as the sum.

Everything here is synthetic data, generated in-notebook, so you can follow along without any external files.

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(7)

# Team A: many low-quality chances
team_a = rng.beta(1.2, 9.0, size=18)

# Team B: fewer chances, but one clear look
team_b = np.concatenate([rng.beta(1.2, 9.0, size=5), [0.62]])

print('Team A xG:', round(team_a.sum(), 2))
print('Team B xG:', round(team_b.sum(), 2))
```

```python
# The totals are close. The experiences were not.
```

## The shape, not the sum

Team A had eighteen pot-shots and not much else. Team B had a handful of moves and one very good chance. The model sees almost identical totals. A person watching the match sees two very different games.

If you wanted to rank "how close was Team B to scoring", the sum is misleading; the *max* is closer to the truth.

```python
pd.DataFrame({
    'team': ['A', 'B'],
    'shots': [len(team_a), len(team_b)],
    'xg_sum': [round(team_a.sum(), 2), round(team_b.sum(), 2)],
    'xg_max': [round(team_a.max(), 2), round(team_b.max(), 2)]
})
```

## Takeaway

When you see a post-match xG line, ask for the distribution, not just the total. The sum is the headline; the shape is the story.
